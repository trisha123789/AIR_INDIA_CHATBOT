```python
from langchain_community.document_loaders import DirectoryLoader
```

```python
loader = DirectoryLoader(
    path = '/content/test',
    glob = '*.pdf',
    loader_cls =PyPDFLoader
)
```

```python
documents = loader.load()
```

```python
context  = "\n\n".join(doc.page_content for doc in documents)
```

```python
!pip install langchain_core
```

```python
from langchain_core.prompts import PromptTemplate
```

```python
prompt = PromptTemplate(
    input_variables = ["context","question"],
    template= """
You are an intelligent assistant trained to understand and provide information based on context provided.
Context :{context}
Question : {question}
your responses should directly address the question using the information from the context,while being clear and concise

    """
)
```

```python
formated_prompt = prompt.format(
    context = context,
    question = "when did air inddia crash 171 happened? "
)
```

```python
result = chat.invoke(formated_prompt)
```

```python
print(result.content)
```
